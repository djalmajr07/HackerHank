```python
# s='HackerRank.com presents "Pythonist 2".'


def swap_case(s):
    new_s = list()
    for char in s:
        if char.islower():
            new_s.append(char.upper())
        elif char.isupper():
            new_s.append(char.lower())
        else: 
            new_s.append(char)
    return "".join(new_s)



swap_case('HackerRank.com presents "Pythonist 2".')
```
