Exercise 2: Clean the dataset and update the CSV file


```python
# The na_values parameter of the read_csv() function is a dictionary that maps column names to a list of values that should be treated as NaN in the resulting DataFrame.
df=pd.read_csv('../dataset/Automobile_data.csv',na_values={
'price':["?","n.a"],
'stroke':["?","n.a"],
'horsepower':["?","n.a"],
'peak-rpm':["?","n.a"],
'average-mileage':["?","n.a"]})
```

```python
mask= (df == '?') | (df == 'n.a')

for col in df.columns[mask.any()]:
    print(col)
```
